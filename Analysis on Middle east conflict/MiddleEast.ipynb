Show number of political event every year around the world

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('middle.xlsx')


```

```python
data = {}
for i in range (2013,2025):
    data[i]=0
```

```python
for i in range(len(df['YEAR'])):
    data[df['YEAR'][i]] += 1
```

```python
keys = list(data.keys())
values = list(data.values())
    
plt.bar(keys, values)
plt.xlabel('Count')
plt.ylabel('Year')
plt.title('Polictical violence by year')
plt.show()
```

```python
import seaborn as sns
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
month = {}
for i in range (1997,2025):
    for x in months:
        month[(i,x)]=0
```

```python
for i in range(len(df['Year'])):
    month[(df['Year'][i],df['Month'][i])] += df['Events'][i]
```

```python
months = [key[1] for key in month.keys()]
event_counts = list(month.values())

# Plotting the bar plot
plt.figure(figsize=(10, 6))
plt.bar(months, event_counts, color='skyblue')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.title('Events per Month')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
month = {}
for i in range (2015,2025):
    for x in months:
        month[(i,x)]=0
```

```python
for i in range(len(df['Year'])):
    if(df['Year'][i])>= 2015:
        month[(df['Year'][i],df['Month'][i])] += df['Events'][i]
```

```python
ye = [2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
event_matrix = np.zeros((len(ye), 12))
for i in range(len(ye)):
    for j in range(12):
        event_matrix[i, j] = month[(ye[i],months[j])]
```

```python
# Create a heatmap
plt.figure(figsize=(11, 8))
sns.heatmap(event_matrix, annot=True, fmt='g',
            xticklabels=months, yticklabels=ye)
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Event Count Heatmap')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
EType={}
```

```python
yr = [2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
```

```python
name = []
```

```python
for x in df['SUB_EVENT_TYPE']:
    if x not in name:
        name.append(x)
    for i in yr:
        EType[(x,i)]= 0
```

```python
for i in range (len(df['SUB_EVENT_TYPE'])):
    EType[(df['SUB_EVENT_TYPE'][i],df['YEAR'][i])] += 1
```

```python
import numpy as np
```

```python
ye = [2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
event_matrix = np.zeros((24,len(ye)))
for j in range(len(ye)):
    for i in range(24):
        event_matrix[i, j] = EType[(name[i],ye[j])]
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(11, 8))
sns.heatmap(event_matrix, annot=True, fmt='g',
            xticklabels=ye, yticklabels=name)
plt.xlabel('Year')
plt.ylabel('Event Type')
plt.title('Event Count Heatmap')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
